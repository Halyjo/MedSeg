```python
def convtransposed_size(insize, filt, pad=0, stride=1, dilation=1, outpad=1):
    n, cin, h, w = insize
    cin, cout, fh, fw = filt
    size = np.array([h, w])
    filtersize = np.array([fh, fw])
    outsize = (size-1)*stride - 2*pad + dilation*(filtersize-1) + outpad + 1
    return (n, cout, *outsize)
```

```python
convtransposed_size((1, 1, 32, 32), (1, 1, 16, 16), stride=16, dilation=1, outpad=0)
```

```python
convtransposed_size((1, 1, 16, 16), (1, 1, 32, 32), stride=(34, 3), dilation=(1,1), outpad=(1,1))
```
